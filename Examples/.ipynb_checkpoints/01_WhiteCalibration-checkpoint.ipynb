```python
## MODIFY
PathToFunctions = 'GitHub/HySE/'
sys.path.append(PathToFunctions)
import HySE
```

# Define Wavelengths used

```python
## Wavelengths as input in the NKT software during acquisition
##                     0.   1.   2.   3.   4.   5.   6.   7
Panel2_Wavelengths = [625, 606, 584, 560, 543, 511, 486, 418]
Panel4_Wavelengths = [646, 617, 594, 576, 569, 526, 503, 478]


Wavelengths_list = np.array(Panel2_Wavelengths+Panel4_Wavelengths)
Wavelengths_list_sorted = np.sort(Wavelengths_list)
```

# Define where the data is

```python
## MODIFY

## Where to save results:
SavingPath = ''

## Path to the main data folder
DataPath_General = ''

## Path to the file containing the dark
DarkPath = DataPath_General+'Dark.mp4'

## Paths to the files containing white calibration
##   Using Mixed wavelengths illumination (HySE) 
##   And Single Wavelength illumination (SingleWav)
White_SingleWav_Path = DataPath_General+'SingleWav_White.mp4'
White_HySE_Path = DataPath_General+'HySE_White.mp4'
```

# Dark

```python
## Import and process the dark:
LongDark = HySE.GetDark_WholeVideo(DarkPath) #, CropImDimensions=CropImDimensions
HySE.PlotDark(LongDark)
np.savez(f'{SavingPath}LongDark.npz', LongDark)
```

# Load Data

```python
## Load the single wavelength data file and identify sweeps:
%matplotlib ipympl
EdgePos_SingleWav = HySE.FindHypercube_RGB(White_SingleWav_Path, Automatic=True, PlateauSize=9, DarkMin=12, MaxPlateauSize=20, SaveFig=False)
```

```python
## Load the mixed wavelengths data file and identify sweeps:
EdgePos_HySE = HySE.FindHypercube_RGB(White_HySE_Path, Automatic=True, PlateauSize=9, DarkMin=12, MaxPlateauSize=20, RedOffset=0, SaveFig=False)
```

# Extract Frames

```python
## Using the previously identified sweep positions
Buffer = 3
Frames_SingleWav, _ = HySE.ComputeHypercube_RGB(White_SingleWav_Path, EdgePos_SingleWav, Buffer=Buffer, BlueShift=1, SaveArray=False)
print(Frames_SingleWav.shape)
```

```python
Buffer = 3
Frames_HySE, _ = HySE.ComputeHypercube_RGB(White_HySE_Path, EdgePos_HySE, Buffer=Buffer, BlueShift=1, SaveArray=False)
print(Frames_HySE.shape)
```

# Dark Subtraction

```python
## SINGLE WAV
## -------

## (Nsweeps, Nwav, Nframes, Y, X, _) = Frames.shape
## BGR 
Frames_SingleWav_Blue = Frames_SingleWav[:,:,:,:,:,0]
Frames_SingleWav_Green = Frames_SingleWav[:,:,:,:,:,1]
Frames_SingleWav_Red = Frames_SingleWav[:,:,:,:,:,2]


## Use NormaliseMixedHypercube to apply dark normalisation:
Frames_SingleWav_BlueD, MaskB = HySE.NormaliseMixedHypercube(Frames_SingleWav_Blue, Dark=LongDark, SaveFigure=False, Plot=False)
Frames_SingleWav_GreenD, MaskG = HySE.NormaliseMixedHypercube(Frames_SingleWav_Green, Dark=LongDark, SaveFigure=False, Plot=False)
Frames_SingleWav_RedD, MaskR = HySE.NormaliseMixedHypercube(Frames_SingleWav_Red, Dark=LongDark, SaveFigure=False, Plot=False)

print(Frames_SingleWav_GreenD.shape)
```

```python
## HySE
## -------

## (Nsweeps, Nwav, Nframes, Y, X, _) = Frames.shape
## BGR 
Frames_HySE_Blue = Frames_HySE[:,:,:,:,:,0]
Frames_HySE_Green = Frames_HySE[:,:,:,:,:,1]
Frames_HySE_Red = Frames_HySE[:,:,:,:,:,2]


## Use NormaliseMixedHypercube to apply dark normalisation:
Frames_HySE_BlueD, MaskB = HySE.NormaliseMixedHypercube(Frames_HySE_Blue, Dark=LongDark, SaveFigure=False, Plot=False)
Frames_HySE_GreenD, MaskG = HySE.NormaliseMixedHypercube(Frames_HySE_Green, Dark=LongDark, SaveFigure=False, Plot=False)
Frames_HySE_RedD, MaskR = HySE.NormaliseMixedHypercube(Frames_HySE_Red, Dark=LongDark, SaveFigure=False, Plot=False)

print(Frames_HySE_GreenD.shape)
```

# Compute Average Per Frame

```python
"""
Sanity check plot - if not straight, issues with the data and/or data extraction

"""

# Frames_SingleWav_GreenD, Frames_SingleWav_RedD
print(Frames_SingleWav_GreenD.shape)
print(Frames_HySE_BlueD.shape)

plt.close()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,5))
test = np.average(np.average(Frames_HySE_Red, axis=3), axis=3)

ax.plot(test[0,:,1], '.-', c='gold')
ax.plot(test[1,:,1], '.-', c='cyan')
ax.plot(test[2,:,1], '.-', c='red')
ax.set_ylim(0,100)
plt.show()
```

```python
def AverageAllWhiteCalibrationFrames(WhiteCalibrationData):
    ## (Nsweeps, Nwavs, Nframes, YY, XX) = WhiteCalibrationData.shape
    Avg = np.average(np.average(np.average(np.average(WhiteCalibrationData,axis=0), axis=1), axis=1), axis=1)
    return Avg
```

```python
Trace_SingleWav_Blue = AverageAllWhiteCalibrationFrames(Frames_SingleWav_BlueD)
Trace_SingleWav_Green = AverageAllWhiteCalibrationFrames(Frames_SingleWav_GreenD)
Trace_SingleWav_Red = AverageAllWhiteCalibrationFrames(Frames_SingleWav_RedD)

Trace_HySE_Blue = AverageAllWhiteCalibrationFrames(Frames_HySE_BlueD)
Trace_HySE_Green = AverageAllWhiteCalibrationFrames(Frames_HySE_GreenD)
Trace_HySE_Red = AverageAllWhiteCalibrationFrames(Frames_HySE_RedD)
```
